```python
import seaborn as sns
```

```python
# Basic plotting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter of total bill vs tip")
plt.show()
```

```python
# line plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line plot of size vs total bill")
plt.show()
```

```python
## Categorical plots
## Bar Plot

sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar plot of day vs total bill")
plt.show()
```

```python
## Box Plot

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box plot of day vs total bill")
plt.show()
```

```python
## Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin plot of day vs total bill")
plt.show()
```

```python
### Histograms

sns.histplot(tips['total_bill'], bins=10)
plt.title("Histogram of total bill")
plt.show()
```
